```python
import pandas as pd
from dataclasses import dataclass
```

```python
xlsx = pd.ExcelFile("NAZ CornDataBook 9.7 09 27.xlsx")
```

```python
customers_demand_data = pd.read_excel(xlsx, 'Customers Demand')
customers_demand_data.head()
```

```python
customers_demand_data.drop_duplicates(subset=["Customer Name","Product Name"], keep="last", inplace=True)
```

```python
customers_demand_data = customers_demand_data.iloc[:, [2,3,5]]
```

```python
customers_demand_data
```

```python
customers_demand_data["Customer Name"].unique()
```

```python
suppliers_capacity_data = pd.read_excel(xlsx, 'Suppliers Capacity')
suppliers_capacity_data.head()
```

```python
suppliers_capacity_data.drop_duplicates(subset=["Supplier Name","Product Name"], keep="last", inplace=True)
```

```python
suppliers_capacity_data = suppliers_capacity_data.iloc[:, [2,4,7]]
```

```python
suppliers_capacity_data
```

```python
production_cost_data = pd.read_excel(xlsx, 'Production Cost')
production_cost_data.head()
```

```python
production_cost_data.drop_duplicates(subset=["Supplier Name","Product Name"], keep="last", inplace=True)
```

```python
production_cost_data = production_cost_data.iloc[:,1:4]
```

```python
production_cost_data
```

```python
@dataclass
class Product:
    name: str
    supplier: str
    cost: float
    
    def __init__(self, name, supplier, cost):
        self.name = name
        self.supplier = supplier
        self.cost = cost
```

```python
products = {}

for prod_name in production_cost_data["Product Name"].unique() :
    products[prod_name] = []
    
products
```

```python
for prod_index, prod_row in production_cost_data.iterrows():
    products[prod_row["Product Name"]].append(Product(prod_row["Product Name"], prod_row["Supplier Name"], prod_row["Production Cost (EXW)"]))
```

```python
products
```

```python
transportation_cost_data = pd.read_excel(xlsx, 'Transportation Cost')
transportation_cost_data.head()
```

```python
@dataclass
class Lane(): 
    customer: str
    supplier: str
    transport_cost: float
    product: Product
        
    def __init__(self, customer, supplier, transport_cost, product):
        self.customer = customer
        self.supplier = supplier
        self.transport_cost = transport_cost
        self.product = product
```

```python
lanes = []

for tran_index, tran_row in transportation_cost_data.iterrows():
    for product in products[tran_row["Product Name"]] :
        if (product.supplier == tran_row["Site From"]) :
            lanes.append(Lane(tran_row["Site To"], tran_row["Site From"], tran_row["Transportation Cost"], product))
```

```python
lanes
```

```python
@dataclass
class Customer:
    name: str
    product_demand: list
    
    def __init__(self, name, product_demand):
        self.name = name
        self.product_demand = product_demand
```

```python
customers_dict = {}

for name in customers_demand_data["Customer Name"].unique() :
    customers_dict[name] = []
```

```python
for i, row in customers_demand_data.iterrows() :
    customers_dict[row["Customer Name"]].append((row["Product Name"], row["Base Demand"]))
```

```python
customers_dict
```

```python
customers = []

for key, value in customers_dict.items():
    customers.append(Customer(key, value))
```

```python
customers
```

```python
@dataclass
class Supplier:
    name: str
    product_capacity: list
    
    def __init__(self, name, product_capacity):
        self.name = name
        self.product_capacity = product_capacity
```

```python
suppliers_dict = {}

for name in suppliers_capacity_data["Supplier Name"].unique() :
    suppliers_dict[name] = []
```

```python
for i, row in suppliers_capacity_data.iterrows() :
    for product in products[row["Product Name"]] :
        if (product.supplier == row["Supplier Name"]) :
            suppliers_dict[row["Supplier Name"]].append((product, row["Capacity (contracted for ABI)"]))
```

```python
suppliers_dict
```

```python
suppliers = []

for key, value in suppliers_dict.items():
    suppliers.append(Supplier(key, value))
```
